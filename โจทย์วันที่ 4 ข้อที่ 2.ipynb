```python
import pandas as pd
import numpy as np
```

```python
#เปลี่ยน Duration >= เป็น 20
a = pd.read_csv("data2.csv")
for x in a.index:
  if a.loc[x, "Duration"] >= 60:
     a.loc[x, "Duration"] = 20
     print(a)
```

```python
#เปลี่ยน Pulse เป็น ทศนิยม
a = pd.read_csv("data2.csv")
a["Pulse"] = a["Pulse"].astype(float)
print(a)
```

```python
#เปลี่ยนที่ที่เป็น NaN เป็นค่าว่า "ว่าง"
a = pd.read_csv("data2.csv")
a.fillna("ว่าง", inplace = True)
print(a)
```
